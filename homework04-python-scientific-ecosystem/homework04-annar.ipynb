### Exercise 3 - Flatten the curve

Copy the function `new_infections(t, k)` from last week's homework (exercise 3) and re-do the exercise using Numpy arrays instead of Python lists.

What needs to be changed in the function `new_infections(t, k)` to make this work?

```python
import math
import numpy as np

P = 1000000
i = 1
k = 3 / (P*10)
t = np.arange(250)
capacity = 10000
```

```python
def exp(k,P,t):
    e = np.exp(-k * P * t)
    return e 

e = exp(k, P, t)

def new_infections(t, k):
    infections = (e* k * P**2 * (-1 + P/i)) / (1 + e * (-1 + P/i))**2
    return infections

infections = new_infections(t, k)
```

```python
plt.plot(t, infections, label = "infection")
plt.xlabel("Time")
plt.ylabel("No new infections")
plt.title("Flatten the curve")
plt.plot ((0,250), (capacity,capacity), label = "Healthcare capacity")
plt.legend()
plt.show
```
